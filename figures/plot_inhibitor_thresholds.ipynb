```python
data_path = '../experimental_data/inhibitor_thresholds/'
osc_thr_wee1 = pd.read_csv(data_path + 'oscil_thresh_wee1_08_10_20.txt', sep=' ')
osc_thr_cdc25 = pd.read_csv(data_path + 'oscil_thresh_cdc25_01_06_21.txt', sep=' ')
per_thr_wee1 = pd.read_csv(data_path + 'per_thresh_wee1_08_10_20.txt', sep=' ')
per_thr_cdc25 = pd.read_csv(data_path + 'per_thresh_cdc25_01_06_21.txt', sep=' ')
print(osc_thr_wee1)
print(per_thr_wee1)
```

```python
osc_yerr = (osc_thr_wee1['Thresh'] - osc_thr_wee1['Thresh_Low'], 
            osc_thr_wee1['Thresh_High'] - osc_thr_wee1['Thresh'])
per_yerr = (per_thr_wee1['Thresh'] - per_thr_wee1['Thresh_Low'], 
            per_thr_wee1['Thresh_High'] - per_thr_wee1['Thresh'])
plt.errorbar(osc_thr_wee1['Inh_Conc'], osc_thr_wee1['Thresh'], yerr=osc_yerr, linestyle='--',
            marker='o', capsize=5, label='Oscillation threshold')
plt.errorbar(per_thr_wee1['Inh_Conc'], per_thr_wee1['Thresh'], yerr=per_yerr, linestyle='--',
            marker='o', capsize=5, label='Period doubling threshold')
plt.ylim([40,100])
lgnd = plt.legend(fontsize=12)
plt.tick_params(labelsize=14)
plt.xlabel('Inhibitor concentration', fontsize=14)
plt.ylabel('Dilution Percentage', fontsize=14)
plt.savefig('./SVGs/Wee1_Thresholds_Fig_4.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
osc_yerr = (osc_thr_cdc25['Thresh'] - osc_thr_cdc25['Thresh_Low'], 
            osc_thr_cdc25['Thresh_High'] - osc_thr_cdc25['Thresh'])
per_yerr = (per_thr_cdc25['Thresh'] - per_thr_cdc25['Thresh_Low'], 
            per_thr_cdc25['Thresh_High'] - per_thr_cdc25['Thresh'])
plt.errorbar(osc_thr_cdc25['Inh_Conc'], osc_thr_cdc25['Thresh'], yerr=osc_yerr, linestyle='--',
            marker='o', capsize=5, label='Oscillation threshold')
plt.errorbar(per_thr_cdc25['Inh_Conc'], per_thr_cdc25['Thresh'], yerr=per_yerr, linestyle='--',
            marker='o', capsize=5, label='Period doubling threshold')
plt.ylim([30,95])
lgnd = plt.legend(fontsize=12)
plt.tick_params(labelsize=14)
plt.xlabel('Inhibitor concentration', fontsize=14)
plt.ylabel('Dilution Percentage', fontsize=14)
plt.savefig('./SVGs/Cdc25_Thresholds_Fig_4.svg', format='svg', bbox_inches='tight')
plt.show()
```
